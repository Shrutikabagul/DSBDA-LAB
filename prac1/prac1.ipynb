```python
#Filling Missing Values:

df["Cabin"]=df["Cabin"].replace(to_replace=np.nan,value="unknown")
```

```python
df["Age"]=df["Age"].interpolate()
```

```python
df.isnull().sum()
```

```python
#Creating Dummy Variables:
quantitative_data=pd.get_dummies(df.Embarked,prefix='Embarked')
```

```python
quantitative_data
```

```python
#Joining Dummy Variables
df=df.join(quantitative_data)
```

```python
df
```

```python
#Dropping Original Categorical Column
df.drop(['Embarked'],axis=1,inplace=True)
```

```python
df
```

```python
def fun1(value):
    if (value == "male"):
        return 1
    elif (value == "female"):
        return 0
    else:
        pass
```

```python
df["Sex"].value_counts()
```

```python
df["Sex"] = df["Sex"].apply(fun1)
```
